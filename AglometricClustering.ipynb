```python
import scipy.cluster.hierarchy as shc
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt1
import pandas as pd
import matplotlib.pyplot as plt2
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans

df = pd.read_csv("agglomerative_clustering_with_labels.csv")
data = df[["feature1", "feature2"]]
print(data)

# -------------------- Dendrogram --------------------
plt.figure(figsize=(10, 6))
plt.title("Hierarchy Clustering")
dendogram = shc.dendrogram(shc.linkage(data, method='ward'))
plt.show()

# -------------------- Agglomerative clustering --------------------
dataTest = AgglomerativeClustering(n_clusters=2, metric="euclidean", linkage="ward")
finalData = dataTest.fit(data)

# -------------------- Agglomerative Scatter plot --------------------
plt1.figure(figsize=(8, 5))
plt1.scatter(data.iloc[:, 0], data.iloc[:, 1],
             c=finalData.labels_, cmap='viridis', marker='o')

plt1.title("Agglomerative Clustering")
plt1.xlabel("Feature 1")
plt1.ylabel("Feature 2")
plt1.show()

# -------------------- K-Means Clustering --------------------
kmeans = KMeans(n_clusters=2, random_state=42, n_init="auto")
kmeans_labels = kmeans.fit_predict(data)   #  FIXED LINE

# -------------------- K-Means Scatter plot --------------------
plt2.figure(figsize=(8, 5))
plt2.scatter(data.iloc[:, 0], data.iloc[:, 1],
             c=kmeans_labels, cmap='plasma', marker='o')

plt2.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],
             s=200, c='black', marker='X', label='Centroids')

plt2.title("K-Means Clustering")
plt2.xlabel("Feature 1")
plt2.ylabel("Feature 2")
plt2.legend()
plt2.show()
```
