```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
x=np.array([[1,2],[1,4],[1,0],[10,4],[10,0],[5,8],[6,7],[4,9]])
kmeans=KMeans(n_clusters=3,random_state=0,n_init="auto")
kmeans.fit(x)
labels=kmeans.labels_
clusters=kmeans.cluster_centers_
print(labels)
new_data=np.array([[0,0],[12,3],[5,5]])
predict=kmeans.predict(new_data)
print(predict)
print(clusters)
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8, label='Data Points')
plt.scatter(clusters[:,0],clusters[:,1],c='purple',marker='X',s=100,alpha=0.8,label="Cluster Centers")
plt.scatter(new_data[:,0],new_data[:,1],c='purple',marker='o',s=100,alpha=0.7,label="New Data Pointa")
plt.title('K Means Clustering')
plt.xlabel('Feature-1')
plt.ylabel('Feature-2')
plt.show()
```

```python
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np


X = np.array([[1, 2], [2, 3], [3, 3], [6, 5], [7, 8], [8, 6]])
y = np.array([0, 0, 0, 1, 1, 1])  # class labels (0 and 1)


model = SVC(kernel='linear')  # kernels: 'linear', 'rbf', 'poly', 'sigmoid'
model.fit(X, y)

X_new = np.array([[2, 2], [5, 4], [7, 7]])
predicted = model.predict(X_new)
print("Predicted classes:", predicted)


plt.scatter(X[:, 0], X[:, 1], c=y, cmap='bwr', label='Training data')
plt.scatter(X_new[:, 0], X_new[:, 1], c=predicted, marker='x', s=100, label='Predictions')


ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

xx, yy = np.meshgrid(np.linspace(xlim[0], xlim[1], 100),
                     np.linspace(ylim[0], ylim[1], 100))
Z = model.decision_function(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)


ax.contour(xx, yy, Z, colors='k', levels=[-1, 0, 1],
           alpha=0.5, linestyles=['--', '-', '--'])
plt.legend()
plt.title("SVM Classification (Linear Kernel)")
plt.show()

```
