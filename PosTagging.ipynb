```python
import spacy
```

```python
nv=spacy.load('en_core_web_sm')
```

```python
doc=nv("Apple is looking to buy Samsung")
print(doc[0].pos_)
print(doc[0].tag_)
spacy.explain("NNP")
for word in doc:
  print(word.text,"------>",word.pos_,word.tag_,spacy.explain(word.tag_))
```

```python
doc1=nv(u"I left the room ")
doc2=nv(u"I to the left of the room")
for word in doc1:
  print(word.text,word.pos_,word.tag_,spacy.explain(word.tag_))
print("---------------Seperate-------------")
for word in doc2:
  print(word.text,word.pos_,word.tag_,spacy.explain(word.tag_))
```
