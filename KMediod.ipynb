```python
import numpy as np
from kmedoids import KMedoids
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs


X, _ = make_blobs(n_samples=300, random_state=40, centers=4)


kmdd = KMedoids(n_clusters=4, random_state=0, init="random", metric='euclidean')
kmdd.fit(X)


labels = kmdd.labels_
medoid_indices = kmdd.medoid_indices_
medoids = X[medoid_indices]


plt.scatter(X[:, 0], X[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8, label='Data Points')


plt.scatter(medoids[:, 0], medoids[:, 1], c='red', marker='X', s=200, label='Medoids')


plt.legend()
plt.grid()
plt.title("K-Medoids Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")


plt.show()
```

```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
x=np.array([[1,2],[1,4],[1,0],[10,4],[10,0],[5,8],[6,7],[4,9]])
kmeans=KMeans(n_clusters=3,random_state=0,n_init="auto")
kmeans.fit(x)
labels=kmeans.labels_
clusters=kmeans.cluster_centers_
print(labels)
new_data=np.array([[0,0],[12,3],[5,5]])
predict=kmeans.predict(new_data)
print(predict)
print(clusters)
plt.scatter(x[:, 0], x[:, 1], c=labels, cmap='viridis', s=50, alpha=0.8, label='Data Points')
```
