```python
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

import numpy as np
x = np.array([[1], [2], [3], [4], [5]])   # shape: (5, 1)
y = np.array([7, 10, 13, 16, 19])         # shape: (5, )
model=LinearRegression()
model.fit(x,y)
X_new = np.array([[6], [7], [8], [9], [10]])
predicted=model.predict(X_new)
print(predicted)
plt.scatter(x, y, color='blue', label='Training data')
plt.scatter(X_new, predicted, color='red', label='Predictions')
plt.plot(X_new, predicted, color='green', linestyle='--', label='Regression line')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Linear Regression Example')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
import matplotlib.pyplot as plt
import numpy as np

# Example binary classification data
x = np.array([[1], [2], [3], [4], [5]])
y = np.array([0, 0, 0, 1, 1])   # categorical classes (0 or 1)

model = LogisticRegression()
model.fit(x, y)

# Predict probabilities and classes for new data
X_new = np.array([[1.5], [2.5], [3.5], [4.5]])
predicted = model.predict(X_new)
probabilities = model.predict_proba(X_new)

print("Predicted classes:", predicted)
print("Predicted probabilities:\n", probabilities)

# Plot
plt.scatter(x, y, color='blue', label='Training data')
plt.scatter(X_new, predicted, color='red', label='Predicted classes')
plt.xlabel('X values')
plt.ylabel('Class (0 or 1)')
plt.title('Logistic Regression Example')
plt.legend()
plt.show()
```
